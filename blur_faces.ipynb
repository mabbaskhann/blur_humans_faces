```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
face_classifier = cv2.CascadeClassifier('models/haarcascade_frontalface_default.xml')
```

```python
img = cv2.imread('images/people.jpg')
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
def blur_faces(fix_img):
    face_rect = face_classifier.detectMultiScale(fix_img)
    for (x,y,w,h) in face_rect:
        crop_img = fix_img[y:y+h, x:x+w]
        blur_img = cv2.GaussianBlur(crop_img,(21,21),150)
        fix_img[y:y+h, x:x+w] = blur_img
        

    return fix_img
```

```python
result = blur_faces(fix_img)
plt.imshow(result)
```
